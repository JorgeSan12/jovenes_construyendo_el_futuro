```python
print('Esta es una calculadora')
print('Elige el tipo de operacion a realizar')
print('1. Suma')
print('2. Resta')
print('3. Multiplicación')
print('4. División')
res = input('¿Que operacion se hara? ')
num1 = float(input("Escribe el primer número: "))
num2 = float(input("Escribe el segundo número: "))
if res == 'suma':
  print('{} + {} = {}'.format(num1 , num2 , num1 + num2))
elif res == 'resta':
 print('{} - {} = {}'.format(num1 , num2 , num1 - num2))
elif res == 'multiplicacion':
 print('{} * {} = {}'.format(num1 , num2 , num1 * num2))
elif res == 'division':
 print('{} / {} = {}'.format(num1 , num2 , num1 / num2))
else:
  print('Comando no valido')
```
