```python
#un programa donde le piden al usuario el limite superior para realizar
#la suma partiendo desde cero y hasta llegar al limite superior, si me 
#dice que el limite superior es 145
#1+2+3+4+...+145
print('Suma que comienza en 0 y termina en el limite indicado')
ls = int(input('Ingrese el limite superior: '))
suma = 0
for i in range(ls + 1):
  suma += i
print(suma)
```

```python
print('Suma de pares que comienza en 1 y termina en el limite indicado')
ls = int(input('Ingrese el limite superior: '))
suma = 0
for i in range(1 , ls + 1):
  if i%2 == 0:
    suma += i
print(suma)
```
